```python
import anthropic
import os
```

```python
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("ANTHROPIC_API_KEY") # save key in another .env file 

print(api_key[:12]) #superbly secret key
```

```python
client = anthropic.Anthropic(api_key=api_key)
```

```python
import anthropic
print(anthropic.__version__)
```

```python
models = client.models.list()

for m in models.data:
    print(m.id)
```

```python
# testing api 
response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=100,# max tokens used in response
    messages=[
        {"role": "user", "content": "Say hello"}
    ]
)

print(response.content[0].text)
```

```python
# the chatbot, asking the question 
def ask_claude(question):

    response = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=200,
        messages=[
            {"role": "user", "content": question}
        ]
    )

    return response.content[0].text
```

```python
print(ask_claude("Explain neural networks simply"))
```
